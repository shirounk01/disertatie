Fetch the information of a problem from a provided link

````python
from bs4 import BeautifulSoup
from curl_cffi import requests
import json
import re
import csv

def parse_html_content(link):

    response = requests.get(
        f"{link}",
        impersonate="safari_ios"
    )

    html_content = response.text

    soup = BeautifulSoup(html_content, 'html.parser')

    # problem difficulty
    problem_difficulty = soup.find("span", {"title": "Difficulty"})
    problem_difficulty = ''.join(re.findall(r'\d+', problem_difficulty.text))

    soup = soup.find('div', class_='ttypography')

    # name
    problem_name = soup.find('div', class_='title').text.strip()

    # notes
    try:
        notes = soup.find('div', class_='note').find_all('p')
        notes += soup.find('div', class_='note').find_all('li')
        notes_text = '\n'.join([p.text.strip() for p in notes])
    except:
        notes_text = ""

    # input spec
    input_spec = soup.find('div', class_='input-specification').find_all('p')
    input_spec += soup.find('div', class_='input-specification').find_all('li')
    input_spec_text = '\n'.join([p.text.strip() for p in input_spec])

    # output spec
    output_spec = soup.find('div', class_='output-specification').find_all('p')
    output_spec += soup.find('div', class_='output-specification').find_all('li')
    output_spec_text = '\n'.join([p.text.strip() for p in output_spec])

    # problem statement
    problem_statement_div = soup.find('div', class_='problem-statement')
    for section_class in ['input-specification', 'output-specification', 'note']:
        section = problem_statement_div.find('div', class_=section_class)
        if section:
            section.decompose()
    problem_statement = problem_statement_div.find_all('p')
    problem_statement += problem_statement_div.find_all('')

    problem_statement_text = '\n'.join([p.text.strip() for p in problem_statement])

    # sample_input
    sample_input = soup.find('div', class_='sample-test').find('div', class_='input').find('pre')
    for even_div in sample_input.find_all('div', class_='test-example-line-even'):
        even_div.insert_after("\n")
    sample_input = sample_input.get_text(strip=True, separator="\n")

    # sample_output
    sample_output = soup.find('div', class_='sample-test').find('div', class_='output').find('pre').text.strip()

    problem_dict = {
        "id": "",
        "problem_link": "",
        "problem_name": problem_name,
        "problem_statement": problem_statement_text,
        "input_spec": input_spec_text,
        "output_spec": output_spec_text,
        "input_sample": sample_input,
        "output_sample": sample_output,
        "problem_notes": notes_text,
        "problem_difficulty": problem_difficulty

    }
    for key in problem_dict:
        problem_dict[key] = re.sub(r'[^\u0000-\u007F]+', '',problem_dict[key].replace('$$$', ''))
    return problem_dict
````

Write the fetched information in a CSV file

````python
input_file_path = "/content/drive/MyDrive/AlphaCodium-main/AlphaCodium-main/my_problems/sample_problems.csv"
output_file_path = "/content/drive/MyDrive/AlphaCodium-main/AlphaCodium-main/my_problems/my_problems.csv"

with open(input_file_path, "r") as csv_file:
    csv_content = list(csv.reader(csv_file))
    for index in range(300, 400):
        problem_link = csv_content[index][1]
        print(problem_link)
        json_obj = parse_html_content(problem_link)
        json_obj["id"] = index
        json_obj["problem_link"] = problem_link
        with open(output_file_path, "a", newline='') as output_file:
            csv_writer = csv.DictWriter(output_file, fieldnames=json_obj.keys())
            if output_file.tell() == 0:
                csv_writer.writeheader()
            csv_writer.writerow(json_obj)
````

Translate the said information into JSON format

````python
import csv
import json

input_file_path = "/content/drive/MyDrive/AlphaCodium-main/AlphaCodium-main/my_problems/my_problems.csv"
output_file_path = "/content/drive/MyDrive/AlphaCodium-main/AlphaCodium-main/my_problems/problems/my_problems_{}.json"

json_dict = {
    "name": None,
    "description": None,
    "public_tests": {
        "input": [None],
        "is_valid_test": None,
        "output": [None]
    }
}

with open(input_file_path, "r") as csv_file:
    csv_content = list(csv.reader(csv_file))
    for index in range(300, 400):
        problem = csv_content[index]
        problem[3] = problem[3].replace("\n", "\n\n")
        problem[8] = problem[8].replace("\n", "\n\n")

        json_dict["name"]=problem[2]
        json_dict["description"]=f"{problem[3]}\n\nInput\n\n{problem[4]}\n\nOutput\n\n{problem[5]}\n\nExample\n\nInput\n\n\n{problem[6]}\n\n\nOutput\n\n\n{problem[7]}\n\nNote\n\n{problem[8]}"
        json_dict["public_tests"]["input"][0]=problem[6]
        json_dict["public_tests"]["output"][0]=problem[7]
        output_file = output_file_path.format(problem[0]).replace(" ", "")
        json.dump(json_dict, open(output_file, "w"))
        print(json.dumps(json_dict, indent=4))
````

Generate code for problems using either one of the said approaches

````python
import openai
from google.colab import userdata
import json
import re
import pandas as pd
import csv
import subprocess

class HandleCodeGenerators:
    read_folder="/content/drive/MyDrive/AlphaCodium-main/AlphaCodium-main/my_problems/problems/my_problems_{}.json"
    read_problems_file = "/content/drive/MyDrive/AlphaCodium-main/AlphaCodium-main/my_problems/my_problems.csv"
    columns = ["id", "problem_name","problem_link","problem_difficulty"]

    def __init__(self):
        openai.api_key = userdata.get('OPEN_AI_CHATGPT_KEY')
        self.problems_data = pd.read_csv(HandleCodeGenerators.read_problems_file, usecols=HandleCodeGenerators.columns)

    def set_problem(self, problem_id):
        try:
            self.problem_file = HandleCodeGenerators.read_folder.format(problem_id)
            with open(self.problem_file, "r") as json_file:
                self.problem_statement = json.load(json_file)
        except:
            self.problem_file = None
            print("Problem not found")

    def parse_python_code(code):
        result = re.findall(r'(?<=```python\n)(.*?)(?=\n```)', code, re.DOTALL)[0]
        return result

    def solve_alphacodium(self):
        result = subprocess.run(['python', '-m', 'alpha_codium.solve_my_problem', '--my_problem_json_file', self.problem_file], stdout=subprocess.PIPE)
        decoded_result = result.stdout.decode('utf-8')
        self.alphacodium_solution = decoded_result.split("testing solution on private tests with prediction:\n")[1].split('\x1b[32m')[0]

    def solve_chatgpt(self):
        completion = openai.chat.completions.create(
            model="gpt-4o-mini",
            messages=[
                {
                    "role": "user",
                    "content": "Solve the following problem. Only write the Python code. {}".format(self.problem_statement),
                },
            ],
        )
        self.chatgpt_solution = HandleCodeGenerators.parse_python_code(completion.choices[0].message.content)

    def get_alphacodium_solution(self):
        return self.alphacodium_solution

    def get_chatgpt_solution(self):
        return self.chatgpt_solution

    def write_spreadsheet(self, tag):
        assert tag in ["chatgpt", "alphacodium"], "tag must be 'chatgpt' or 'alphacodium'"

        output_file = "/content/drive/MyDrive/AlphaCodium-main/AlphaCodium-main/my_problems/results_{}.csv".format(tag)
        code = self.chatgpt_solution if tag == "chatgpt" else self.alphacodium_solution

        json_obj = {col:None for col in HandleCodeGenerators.columns+["code"]}
        problem = self.problems_data.loc[self.problems_data['problem_name'] == self.problem_statement["name"]]

        json_obj["id"] = int(problem["id"].values[0])
        json_obj["problem_name"] = problem["problem_name"].values[0]
        json_obj["problem_link"] = problem["problem_link"].values[0]
        json_obj["problem_difficulty"] = int(problem["problem_difficulty"].values[0])
        json_obj["code"] = code
        print(code)

        with open(output_file, "a", newline='') as output_file:
            csv_writer = csv.DictWriter(output_file, fieldnames=json_obj.keys())
            if output_file.tell() == 0:
                csv_writer.writeheader()
            csv_writer.writerow(json_obj)


for index in range(300, 400):
    x = HandleCodeGenerators()
    x.set_problem(index)
    x.solve_chatgpt()
    x.write_spreadsheet("chatgpt")
````

Compare provided and expected output for the generated code

````python
import subprocess
import pandas as pd
import csv
from IPython.display import clear_output

input_output_file =  "/content/drive/MyDrive/AlphaCodium-main/AlphaCodium-main/my_problems/my_problems.csv"
code_file = "/content/drive/MyDrive/AlphaCodium-main/AlphaCodium-main/my_problems/results_alphacodium.csv"
output_file_path = "/content/drive/MyDrive/AlphaCodium-main/AlphaCodium-main/my_problems/scores_alphacodium.csv"

iof = pd.read_csv(input_output_file)
cf = pd.read_csv(code_file)
test_name = "test.py"

index = 1

while index < 400:
    iof_data = iof.loc[iof["id"] == index]
    cf_data = cf.loc[cf["id"] == index]
    expected_input, expected_output, code, difficulty = iof_data["input_sample"].values[0], iof_data["output_sample"].values[0], cf_data["code"].values[0], cf_data["problem_difficulty"].values[0]

    f = open(test_name, "w")
    f.write(code)
    f.close()

    try:
        result = subprocess.run(
            ["python", test_name],
            input=expected_input,
            capture_output=True,
            text=True,
            timeout = 5
        )
        actual_output = result.stdout.rstrip()
        valid_answer = expected_output in actual_output
    except:
        actual_output = "\\"
        valid_answer = "TIMEOUT"

    print(expected_output, actual_output, valid_answer, sep="\n\n")

    json_obj = {}
    json_obj["id"] = index
    json_obj["difficulty"] = difficulty
    json_obj["expected_output"] = expected_output
    json_obj["actual_output"] = actual_output
    json_obj["first_test_passed"] = valid_answer

    input()
    clear_output()

    with open(output_file_path, "a", newline='') as output_file:
        csv_writer = csv.DictWriter(output_file, fieldnames=json_obj.keys())
        if output_file.tell() == 0:
            csv_writer.writeheader()
        csv_writer.writerow(json_obj)

    index += 1
````
